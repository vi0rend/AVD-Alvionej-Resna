Menghapus Missing Values

```python
df.fillna({
    'Staff with PhD': df['Staff with PhD'].median(),
    'International Students': df['International Students'].median(),
    'International Faculty': df['International Faculty'].median()
}, inplace=True)
```

```python
df[df.duplicated()]
```

Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Distribusi Peringkat Universitas

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Rank'], bins=20, kde=True, color='pink')
plt.title('Distribusi Peringkat Universitas di Asia (QS Rankings 2024)')
plt.xlabel('Peringkat Universitas')
plt.ylabel('Jumlah Universitas')
plt.show()
```

### **Penjelasan Histogram Distribusi Peringkat Universitas**  
Histogram di atas menunjukkan distribusi peringkat universitas di Asia berdasarkan QS Rankings 2024. Dari visualisasi ini, terlihat bahwa jumlah universitas tidak tersebar merata di setiap rentang peringkat. Sebagian besar universitas berada di peringkat menengah hingga bawah, sementara hanya sedikit universitas yang masuk dalam peringkat teratas. Pola ini mengindikasikan bahwa kompetisi untuk berada di peringkat atas sangat ketat, sedangkan di peringkat lebih rendah jumlah universitas jauh lebih banyak.  

Garis KDE (Kernel Density Estimation) dalam histogram memberikan gambaran yang lebih halus tentang distribusi data. Jika histogram memiliki puncak tertentu, artinya banyak universitas berkumpul dalam kisaran peringkat tersebut. Sebaliknya, distribusi yang landai menunjukkan bahwa universitas tersebar lebih merata.  

### **Insight:**  
- Sebagian besar universitas berada di peringkat menengah-bawah, menunjukkan ketatnya persaingan di peringkat atas.  
- Universitas di peringkat atas cenderung mempertahankan posisinya dengan sumber daya lebih baik.  
- Peluang bagi universitas di peringkat menengah untuk naik dengan memperbaiki faktor akademik utama.  

###  **Tindakan yang Dapat Diambil:**  
- Universitas peringkat bawah → Tingkatkan publikasi ilmiah & rasio fakultas-mahasiswa.  
- Universitas peringkat menengah → Fokus pada kerja sama internasional & menarik mahasiswa asing.  
- Universitas peringkat atas → Pertahankan posisi dengan riset berdampak tinggi & kemitraan industri.  





Hubungan antara Academic Reputation dan Peringkat Universitas

-Universitas dengan peringkat rendah cenderung memiliki reputasi akademik yang lebih lemah.  
-Beberapa universitas dengan reputasi sedang masih bisa naik peringkat dengan strategi yang tepat.  

### **Tindakan yang Dapat Diambil:**  
-Universitas harus meningkatkan reputasi akademi dengan meningkatkan kualitas pengajaran dan penelitian.  
-Fokus pada publikasi ilmiah dan kerja sama internasional untuk memperkuat kredibilitas akademik.

Tren International Students terhadap Peringkat Universitas

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['Rank'], y=df['International Students'].values, marker='o', color='pink')
plt.title('Tren International Students terhadap Peringkat Universitas')
plt.xlabel('Peringkat Universitas')
plt.ylabel('Jumlah International Students')
plt.grid()
plt.show()
```

### **Penjelasan Line Chart**  
Grafik menunjukkan tren jumlah International Students terhadap Peringkat Universitas. Terlihat bahwa universitas dengan peringkat lebih baik cenderung memiliki lebih banyak mahasiswa internasional dibandingkan universitas dengan peringkat lebih rendah.  

### **Insight Singkat:**  
-Universitas dengan peringkat tinggi lebih menarik bagi mahasiswa internasional, kemungkinan karena kualitas pendidikan dan reputasi global mereka.  
-Universitas dengan jumlah mahasiswa internasional rendah cenderung memiliki peringkat lebih rendah, mungkin karena kurangnya daya tarik global atau kerja sama internasional.  

### **Tindakan yang Dapat Diambil:**  
-Universitas harus meningkatkan daya tarik bagi mahasiswa internasional dengan menawarkan program internasional dan beasiswa.  
-Kerja sama dengan universitas luar negeri dan program pertukaran mahasiswa dapat meningkatkan keberagaman internasional.  
-Calon mahasiswa dapat mempertimbangkan universitas dengan banyak mahasiswa internasional, karena biasanya lebih inklusif dan memiliki jaringan global yang lebih luas.
